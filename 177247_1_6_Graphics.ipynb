# Построение графиков

```python
import matplotlib.pyplot as plt
import numpy as np
```

Построим простую линейную зависимость, дадим графику название, подпишем оси. Перед этим создадим значения для переменных х и у через функцию linespace из библиотеки numpy.

```python
# Независимая (x) и зависимая (y) переменные
x = np.linspace(0, 20, 40)
y = x

# Построение графика
plt.title('Линейная зависимость y = x') # заголовок
plt.xlabel('x') # ось абсцисс
plt.ylabel('y') # ось ординат
plt.grid() # включение отображение сетки
plt.plot(x, y) # построение графика
```

Можем задать вид и цвет соединительной линии, изменив последнюю строчку кода:

```python
plt.plot(x, y, 'g*') # построение графика
```

# Несколько графиков на одном поле

Добавим на предыдущий график квадратичную зависимость y от х:

```python
# Линейная зависимость
x = np.linspace(0, 20, 40)
y1 = x

# Квадратичная зависимость
y2 = [i**2 for i in x]

# Построение графика
plt.title('Зависимости: y1 = x, y2 = x^2') # заголовок
plt.xlabel('x') # ось абсцисс
plt.ylabel('y1, y2') # ось ординат
plt.grid() # включение отображение сетки
plt.plot(x, y1, x, y2) # построение графика
```

В данном случае функции plot() последовательно передаются массивы независимых и зависимых переменных. Важно, чтобы массивы независимых переменных были одни и те же!

# Отображение графиков на разных полях

```python
# Линейная зависимость
x = np.linspace(0, 20, 40)
y1 = x

# Квадратичная зависимость
y2 = [i**2 for i in x]

# Первый график
plt.figure(figsize=(9, 9))
plt.subplot(2, 1, 1)
plt.plot(x, y1) # построение графика
plt.title('Зависимости: y1 = x, y2 = x^2') # заголовок
plt.ylabel('y1', fontsize=14) # ось ординат
plt.grid(True) # включение отображение сетки

#Второй график
plt.subplot(2, 1, 2)
plt.plot(x, y2) # построение графика
plt.xlabel('x', fontsize=14) # ось абсцисс
plt.ylabel('y2', fontsize=14) # ось ординат
plt.grid(True) # включение отображение сетки
```

Здесь мы применили новые функции.

  Функция figure() предназначена для установки глобальных параметров отображения графиков. Ее аргументом является кортеж, который определяет размер общего поля.

  Функция subplot() служит для определения расположения поля с графиком. Существуют различные способы задания областей для вывода графиков.

 В данном примере мы использовали вариант, где требуется передача трех аргументов: первый - количество строк, второй - количество столбцов в формируемом поле, третий - индекс (номер поля, расчет идет сверху вниз, слева направо).

Помимо остальных, уже известных функций, использовали параметр fontsize для определения размера шрифта для подписи осей.

# Построение графика для нечисловых данных

Ранее мы создавали графики для данных числового типа, где были как зависимые, так и независимые переменные. Однако на практике часто приходится иметь дело с данными, которые не являются числовыми, такими как имена людей, названия городов и т.д. Давайте построим диаграмму, которая будет отображать количество товара в магазине.

```python
# Создаем списки овощей и их количества
vegetables = ['картофель', 'томат', 'огурец', 'свекла', 'баклажан']
counts = [94, 73, 109, 88, 74]

#Строим график
plt.bar(vegetables, counts)
plt.title('Овощи!')
plt.xlabel('Овощи')
plt.ylabel('Количество')
```

Для построения такого графика используется функция bar() – столбчатая диаграмма, где по оси х отображаются все возможные категории наших данных, а по оси у – их числовые значения (количество, стоимость, размер и тд.).

# Построение графиков

В основе построения изображения в pyplot лежит фигура (Figure), к которой можно добавить один или несколько графиков, осей, надписей и других элементов. Для создания графика используется команда plot().

Если в качестве аргумента функции передать список элементов, то они будут отображены на оси у, а на оси х – их индексы.

```python
plt.plot([2, 12, 7, 9, 2, 5])
```

Для того чтобы задать значения для оси х, необходимо указать два списка в качестве аргументов функции.

```python
plt.plot([1, 3, 6, 9, 12, 15], [2, 12, 7, 9, 2, 5])
```

Функция plot() позволяет строить как линейные графики, так и точечные. Для этого необходимо указать тип маркера.

```python
plt.plot([1, 3, 6, 9, 12, 15], [2, 12, 7, 9, 2, 5], 'ro')
```

Также, можно построить круговую диаграмму. Для этого необходимо подать два списка, первый – размеры областей круга, второй – подписи (опционально).

```python
plt.pie([2, 12, 7, 9, 2, 5], labels=['a', 'b', 'c', 'd', 'e', 'f'])
```

# Работа с легендой графика

Для отображения легенды используется функция legend(). Внутри функции можно указать соответсвии линий и меток для них.

```python
x = [1, 5, 10, 15, 20]
y1 = [2, 1, 2, 5, 11]
y2 = [2, 4, 5, 5, 12]

line1, = plt.plot(x, y1, 'o-b', label='line 1')
line2, = plt.plot(x, y2, 'o-.g', label='line 1')
plt.legend((line1, line2), ['L1', 'L2'])
```

Расположение легенды на графике определяется параметром loc, который модет принимать следующие значения.
