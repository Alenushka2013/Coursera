# Понятие метрики

**Метрика** – критерий качества модели.

Для **задачи классификации** существует ряд метрик, которые позволяют оценить качество построенной модели-классификатора. Примером простейшей метрики классификации является Accuracy – точность отнесения объекта к определенному классу.

Для **задачи регрессии** существует ряд других метрик, также предназначенных для оценки качества построенных моделей. Например, часто используется метрика MSE (средний квадрат ошибок), которая показывает, как близко смоделированные значения признака находятся по отношению к фактическим значениям.

# Метрики бинарной классификации

**Бинарная классификация**  (двоичная классификация) (Binary classification) – отнесение объекта к одному из двух классов.

Исторически задача бинарной классификации – это задача об *обнаружении чего-то редкого* в большой совокупности объектов, например, поиск человека, больного туберкулёзом, по флюорографии.

**Метка класса**   (маркер класса) (class label, marker) – число, характеризующее объект и указывающее, к какому классу относится данный объект.

Например, имеются три объекта. Первый объект имеет метку класса «0», второй и третий – имеют метку класса «1». Это означает, что первый объект относится к классу «0», а второй объект и третий объект – к классу «1».

**«Положительный» класс** – класс, который представляет для исследователя интерес; класс, который соответствует искомому объекту.

**«Положительная» метка** – метка, указывающая, что объект относится к положительному классу.

**«Отрицательный» класс** – оставшийся класс, который не представляет для исследователя интерес; класс, который не соответствует искомому объекту.

**«Отрицательная» метка** – метка, указывающая, что объект относится к отрицательному классу.

К метрикам бинарной классификации относят следующие метрики:

1.    Accuracy.

2.    Error rate.

3.    Precision.

4.    Recall.

5.    F1.

6.    Fβ​.

1. **Accuracy** (Доля правильных ответов) – доля объектов, для которых правильно был предсказан класс (доля правильных классификаций):

Accuracy = T / N               (1)

где T  – число объектов, для которых правильно был предсказан класс,

 N – число всех объектов.




2. **Error rate** (Частота ошибок) – доля ошибочных классификаций:

1 - Accuracy  (2)

Недостатки Accuracy:

  - не учитывает **дисбаланс классов** (например, если 99% объектов имеют метку второго класса, то классификатор, всегда указывающий на второй класс, будет иметь высокое значение **Accuracy**);
  - не учитывает **цену ошибки** (например, при постановке неправильного диагноза).
  -  Применение **Accuracy**: когда классы имеют одинаковое значение.
  - При бинарной классификации для каждого объекта в выборке возможно 4 ситуации:
  - **True Positive** (TP) (правильно положительный результат) – классификатор предсказал правильно (T) положительную метку (P).
  - **False Positive** (FP) (ложно положительный результат) – классификатор предсказал неправильно (F) положительную метку (P).
  - **True Negative** (TN) (правильно отрицательный результат) – классификатор предсказал правильно (T) отрицательную метку (N).
  -** False Negative** (FN) (ложно отрицательный результат) – классификатор предсказал неправильно (F) отрицательную метку (N).

Метрика *Accuracy* на практике почти не используется. Рассмотрим ситуацию, когда положительный класс это событие редкое. Возьмем в качестве примера поисковую систему – в нашем хранилище хранятся миллиарды документов, а релевантных к конкретному поисковому запросу – несколько единиц. Благодаря большому дисбалансу, Accuracy классификатора, объявляющего все документы нерелевантными, будет близка к единице, при этом высокое значение метрики будет обеспечено членом TN, в то время как для пользователей более важен высокий TP (см. формулу 3).

Поэтому в случае ассиметрии классов, следует использовать метрики, которые не учитывают TN и ориентируются на TP (метрики *Precision, Recall*).

3. **Precision** (Точность) – доля правильно предсказанных положительных объектов среди всех объектов, отнесенных моделью к положительному классу:

image.png

где   TP – объекты, распознанные классификатором как положительные, образующие положительный класс, который сформировал классификатор.

Она отражает качество распознавания положительных объектов, не учитывая размер истинно положительного класса. Например, если всего объектов 20, положительный класс содержит 15 объектов, 10 истинно положительных объектов были отнесены к этому классу и 0 истинно ложных, то Precision = 1.

Вернемся к примеру с документами: метрика Precision показывает долю истинно релевантных документов среди всех документов, отнесенных классификатором к положительному классу, то есть к классу «релевантные документы». Чем меньше ложноположительных классификаций (FP) будет допускать модель, тем выше будет её Precision.

4. **Recall** (Полнота) – доля правильно предсказанных положительных объектов среди всех объектов положительного класса:

image.png

где   TP – объекты, являющиеся истинно положительными, образующие истинный положительный класс.

Для примера с документами: метрика *Recall* показывает долю найденных документов из всех релевантных (истинно релевантных). Чем меньше ложно отрицательных классификаций (FN), тем выше Recall модели.

Рассмотрим другой пример, связанный с предсказанием злокачественных опухолей:

  - *Precision* покажет, сколько из предсказанных моделью злокачественных опухолей действительно являются злокачественными.
  - *Recall* покажет, сколько из всех злокачественных опухолей удалось выявить.

Модели удобнее сравнивать, когда их качество характеризуется одним числом (одной метрикой). *F1* объединяет метрики *Precision* и *Recall*.

# Метрики многоклассовой классификации

**Многоклассовая классификация** (мультиклассовая классификация) (Multiclass, multinomial classification) – отнесение объекта к одному из множества классов (число классов более двух).

Для многоклассовой классификации расчет метрик усложняется.

**Задача многоклассовой классификации** представляется как совокупность из k задач бинарной классификации (где k – число классов).

Каждая задача бинарной классификации заключается в отделении класса i от остальных классов (i = 1, …, k). Для каждой такой задачи можно рассчитать матрицу ошибок. Таким образом, в итоге будет получено k матриц ошибок.

Получение итогового значения метрики осуществляется одним из двух способов:

  - **Микроусреднение.**
  - **Макроусреднение.**

**Микроусреднение** (Micro-averaging) – построение усредненной матрицы ошибок.

В этом случае рассчитываются средние значения элементов k матриц ошибок, т.е. средние значения TP, FN, FP, TN по всем матрицам. Далее по усредненным значениям TP, FN, FP, TN строится новая матрица ошибок (усредненная матрица ошибок), на основе которой рассчитываются метрики .

При микроусреднении вклад маленького класса в общую метрику будет незаметным (несущественным) по следующей причине.

При расчете итоговых Precision и Recall используются усредненные значения TP, FN, FP. Например, имеется 1000 объектов, 9 соразмерных классов и 1 маленький класс. Если класс маленький (содержит малое число объектов, например, 10), то его величины TP, FN будут также маленькими (TP + FN = 10, т.е. TP ≤ 10 и FN ≤ 10). Величина FP скорее всего также будет маленькой, так как частое предсказание классификатором редкого класса повышает вероятность ошибки. Таким образом, величины TP, FN, FP будут малыми, и только величина TN большой (большинство объектов не принадлежит этому классу). Так как величины TP, FN, FP для малого класса будут малыми, то их вклад в средние величины TP, FN, FP по всем классам будет низким.

Допустим, значения TP, FN, FP для малого класса составляют порядка 10. Тогда как значения TP, FN, FP для любого другого класса составляют порядка 100. Средние значения TP, FN, FP для 10 классов в этом случае будут равны примерно 91, что скорее характеризует 9 соразмерных классов (значение 91 близко к 100), чем маленький класс (значение 91 существенно отличается от 10).

# Метрики регрессии

**Модель классификации** моделирует (предсказывает) метку класса, то есть некоторое категориальное, дискретное значение зависимой переменной y с учетом значений факторных переменных x.

**Модель регрессии** моделирует (предсказывает) вещественное число, непрерывную величину зависимой переменной y с учетом значений факторных переменных x.

Наиболее простой и распространенной формой регрессии является линейная регрессия.
