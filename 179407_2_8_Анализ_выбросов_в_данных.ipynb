```python
import pandas as pd
from etna.datasets import generate_ar_df  # генератор даних
from etna.datasets.tsdataset import TSDataset # перетворювач із Dataframe в Dataset
```

```python
# Створення часового ряду
df = generate_ar_df(periods = 100, freq = "D", random_seed = 13, start_time = '2022-01-01') # freq = "D"  - денний
# Перетворення даних в об'єкт TSDataset
df_ts_format = TSDataset.to_dataset(df)
ts = TSDataset(df_ts_format, "D")
```

```python
ts.plot()
```

```python
# Проведем вирівнювання/вінзоризацію/winzorize
from scipy.stats import mstats
df['target'] = mstats.winsorize(df['target'], limits = [0, 0.1]) # 10 %, зверху 90 - процентіль
df_ts_format = TSDataset.to_dataset(df)
ts = TSDataset(df_ts_format, "D")
```
